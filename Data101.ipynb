####Quiz6 แสดง rows ทั้งหมดที่ผู้ต้องสงสัยมีอายุ 10-17 ปี

````python
SS_warwick['Age range'] == '10-17'
````

````python
SS_warwick[SS_warwick['Age range'] == '10-17']
````

####เด็กเชื้อชาติไหนถูกเรียกค้นเยอะที่สุด

````python
SS_warwick[SS_warwick['Age range'] == '10-17'].groupby('Self-defined ethnicity').count()
````

####เด็กแต่เชื้อชาติไหนถูกเรียกค้นจากตำรวจเชื้อชาติไหน

````python
SS_warwick[SS_warwick['Age range'] == '10-17'].groupby(['Self-defined ethnicity','Officer-defined ethnicity']).count()
````

####Quiz7 เปรียบเทียบว่าตำรวจที่เป็น Asian และ White ชอบเรียกค้นคนอายุเท่าไหรและเชื้อชาติไหน ต่างกันอย่างไร

````python
SS_warwick[SS_warwick['Officer-defined ethnicity'].isin(['Asian', 'White'])].groupby(['Officer-defined ethnicity','Self-defined ethnicity','Age range']).count()
````
