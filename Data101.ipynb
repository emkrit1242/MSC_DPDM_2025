Drop เฉพาะ column ที่มี missing มากกว่า 50%

````python
# Calculate the percentage of missing values for each column
missing_percentages = SS_warwick.isnull().mean()

# Identify columns with more than 50% missing values
columns_to_drop = missing_percentages[missing_percentages > 0.5].index

# Drop the identified columns
SS_warwick_dropped = SS_warwick.drop(columns=columns_to_drop)

# Display the shape of the new DataFrame to show the effect of dropping columns
display(SS_warwick_dropped.shape)

````

````python
SS_warwick_dropped.dropna()
````

Drop เฉพาะ data ที่มี missing เฉพาะ column ที่สนใจ

````python
SS_warwick_dropped.dropna(subset=['Gender','Age range','Self-defined ethnicity','Officer-defined ethnicity','Object of search'])
````
